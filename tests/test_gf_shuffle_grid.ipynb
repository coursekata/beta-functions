# Test: gf_shuffle_grid()

Tests for both categorical X and quantitative X scenarios.

```r
source("gf_lm_cat.R")
source("gf_coef.R")
source("gf_shuffle_grid.R")
library(coursekata)
library(ggformula)
```

## Categorical X tests

```r
# Test 1: default 3x3, random real panel, jitter
# Each run should place real data in a different (random) panel
options(repr.plot.width = 9, repr.plot.height = 9)
gf_shuffle_grid(Height ~ Sex, data = Fingers, plot = "jitter")
```

```r
# Test 2: pin real data to panel 1, reveal label
# Panel 1 (top-left) should say "real data"
gf_shuffle_grid(Height ~ Sex, data = Fingers, plot = "jitter",
                real_data = 1, reveal = TRUE)
```

```r
# Test 3: 2x2 grid, real data at panel 3, reveal
# Panels: 1 2
#         3 4  -> panel 3 (bottom-left) is real
options(repr.plot.width = 6, repr.plot.height = 6)
gf_shuffle_grid(Height ~ Sex, data = Fingers, plot = "jitter",
                nrow = 2, ncol = 2, real_data = 3, reveal = TRUE)
```

```r
# Test 4: 3-level categorical X (later_anxiety ~ condition from er dataset)
# Group-mean segments + b0/b1/b2 arrows should appear on each panel
options(repr.plot.width = 9, repr.plot.height = 9)
gf_shuffle_grid(later_anxiety ~ condition, data = er, plot = "jitter",
                real_data = 5, reveal = TRUE, seed = 1)
```

```r
# Test 5: default — show_model = TRUE, show_coef = FALSE
# Should show group-mean segments but NO b0/b1 arrows
gf_shuffle_grid(Height ~ Sex, data = Fingers, plot = "jitter", seed = 7)
```

```r
# Test 5b: show_model = FALSE — bare jitter, no overlay at all
# (show_coef stays FALSE by default, so nothing is added)
gf_shuffle_grid(Height ~ Sex, data = Fingers, plot = "jitter",
                show_model = FALSE, seed = 7)
```

```r
# Test 5c: show_model = TRUE, show_coef = TRUE — full overlay (segments + b0/b1 arrows)
gf_shuffle_grid(Height ~ Sex, data = Fingers, plot = "jitter",
                show_model = TRUE, show_coef = TRUE, seed = 7)
```

## Quantitative X tests

```r
# Test 6: continuous X, default 3x3 point plot
# Each panel should show a regression line + rise-over-run b1 annotation
gf_shuffle_grid(Thumb ~ Height, data = Fingers, plot = "point", seed = 42)
```

```r
# Test 7: 2x3 grid, continuous X, real data at panel 4
options(repr.plot.width = 9, repr.plot.height = 6)
gf_shuffle_grid(Thumb ~ Height, data = Fingers, plot = "point",
                nrow = 2, ncol = 3, real_data = 4, reveal = TRUE, seed = 99)
```

```r
# Test 8: continuous X with jitter (e.g. age ~ year, or just Thumb ~ Height with jitter)
options(repr.plot.width = 9, repr.plot.height = 9)
gf_shuffle_grid(Thumb ~ Height, data = Fingers, plot = "jitter",
                width = 0.5, seed = 11)
```

## Guard / edge case tests

```r
# Test 9: too many panels (6x5 = 30 > 25) → should error
tryCatch(
  gf_shuffle_grid(Height ~ Sex, data = Fingers, nrow = 6, ncol = 5),
  error = function(e) cat("Expected error:", conditionMessage(e), "\n")
)
```

```r
# Test 10: real_data out of range → should error
tryCatch(
  gf_shuffle_grid(Height ~ Sex, data = Fingers, nrow = 3, ncol = 3, real_data = 99),
  error = function(e) cat("Expected error:", conditionMessage(e), "\n")
)
```

```r
# Test 11: seed reproducibility — two calls with same seed should look identical
options(repr.plot.width = 9, repr.plot.height = 9)
gf_shuffle_grid(Height ~ Sex, data = Fingers, seed = 123)
```

```r
# Re-run test 11 with same seed — should be identical
gf_shuffle_grid(Height ~ Sex, data = Fingers, seed = 123)
```
